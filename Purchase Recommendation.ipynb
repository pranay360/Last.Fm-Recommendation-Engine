```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
import json
```

```python
df1 = pd.read_csv("LastFM_Matrix.csv")
user_rec=df1
```

```python
df2 = df1.drop("user",axis =1)
```

```python
distance_matrix = cosine_similarity(df2.T)
```

```python
distDF = pd.DataFrame(distance_matrix, columns=(df2.columns), index=(df2.columns))
distDF.head()
```

```python
model = NearestNeighbors(n_neighbors=distDF.shape[0])
model.fit(distDF)
nearest_neighbours= model.kneighbors(distDF, return_distance=True)
```

```python
NNDF1 = pd.DataFrame(nearest_neighbours[0],index=distDF.index)
NNDF1=NNDF1[NNDF1.columns[1:11]]
```

```python
NNDF = pd.DataFrame(distDF.columns[nearest_neighbours[1]], index=distDF.index)
top10=NNDF[NNDF.columns[1:11]]
top10.head(10)
```

```python
def collaborativeVal(purchased, distance):
   return sum(purchased*distance)/len(purchased)
```

```python
user_rec=user_rec.astype(float)
for i,row in user_rec.iterrows():
    for item in distDF.columns:
        if row[item]==1:
            user_rec.set_value(i, item, 0)
        else:
            purchased=df2.ix[i][list(top10[top10.index==item].values[0])]
            neighbour_similarities=NNDF1[NNDF1.index==item].values[0]
            user_rec.set_value(i, item, collaborativeVal(purchased,neighbour_similarities))
```

```python
rcmd={}
for i,row in user_rec.iterrows():
    rcmd[int(row.user)]={}
    rcmd[int(row.user)]['Recommendation']=' , '.join(list(user_rec.ix[i][distDF.columns].sort_values(ascending=False)[:5].index))
    rcmd[int(row.user)]['Purchased']=' , '.join(list(df2.ix[i][df2.ix[i][df2.columns]==1].index))
```

```python
for k in sorted(rcmd):
   print "User",k,':',json.dumps(rcmd[k],indent=2)
```
